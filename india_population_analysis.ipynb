```python
import pandas as pd
import matplotlib.pyplot as plt

#Read a csv
data = pd.read_csv('/home/aman/Documents/Data Analysis/World Population/world_population_data.csv')

#Adding difference in population in the mention consecutive years and taking the average diff per year
data["2023-2022"] = (data["2023 population"] - data["2022 population"])//1
data["2022-2020"] = (data["2022 population"] - data["2020 population"])//2
data["2020-2015"] = (data["2020 population"] - data["2015 population"])//5
data["2015-2010"] = (data["2015 population"] - data["2010 population"])//5
data["2010-2000"] = (data["2010 population"] - data["2000 population"])//10

top_20 = data.sort_values(by='2023 population', ascending=False).head(20)

#Filtering India Data
india_population = top_20.loc[top_20['country'] == 'India', ['2023-2022', '2022-2020', '2020-2015', '2015-2010', '2010-2000']]

#Transpose the DataFrame to have years as rows and population as columns
india_population = india_population.T

# Plot a bar chart
ax = india_population.plot(kind='bar', title='Average difference in Population of India by Year')
plt.xlabel('Year')
plt.ylabel('Avg diff/year in a given time period')

# Annotate bars with their respective values
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

plt.show()
```
